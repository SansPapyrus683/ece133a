# newton's method

```python
import numpy as np
import matplotlib.pyplot as plt
```

## defining $n(t)$ and some stuff

```python
def n(t):
    return 10 * t * np.exp(-2 * t) + np.exp(-t)


def diff_n(t):
    e_neg2t = np.exp(-2 * t)
    return 10 * e_neg2t - 20 * t * e_neg2t - np.exp(-t)


def ddiff_n(t):
    # courtesy of wolfram elpha
    return np.exp(-2 * t) * (40 * (t - 1) + np.exp(t))
```

## part a

```python
vals = []
target = 2
for start in [0, 1]:
    at = start

    while abs(n(at) - target) > 1e-5:
        at -= 1 / diff_n(at) * (n(at) - target)
    vals.append(at)
vals = np.array(vals)
```

lemme just also plot them to see if they're any good

```python
t = np.linspace(0, 3, 1000)
plt.plot(t, n(t))
plt.scatter(vals, n(vals), color="orange")
plt.xlabel("t")
plt.xlim(0, 3)
plt.ylabel("n(t)")
```

## part b

i just tried to find the point where the derivative was zero and went from there

```python
best_t = 0
for _ in range(100):
    best_t -= 1 / ddiff_n(best_t) * diff_n(best_t)
```

```python
t = np.linspace(0, 3, 1000)
plt.plot(t, n(t))
plt.scatter([best_t], [n(best_t)], color="orange")
plt.xlabel("t")
plt.xlim(0, 3)
plt.ylabel("n(t)")
```

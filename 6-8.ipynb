# circle fitting

```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
```

## data retrieval

```python
data = scipy.io.loadmat("data/circlefit.mat")

u: np.ndarray = data["u"].reshape(-1)
v: np.ndarray = data["v"].reshape(-1)
```

## problem setup

```python
a = np.vstack([-2 * u, -2 * v, np.ones(len(u))]).T
b = -u ** 2 - v ** 2
```

## solving and display

```python
u_c, v_c, w = np.linalg.lstsq(a, b)[0]
r = np.sqrt(u_c ** 2 + v_c ** 2 - w)
```

```python
fig, ax = plt.subplots()

circ = plt.Circle((u_c, v_c), r, fill=False, label="fitted circle")
ax.add_artist(circ)
ax.scatter(u, v, label="raw data", marker=".")
ax.set_aspect("equal")
ax.legend()
```

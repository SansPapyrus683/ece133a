# logistic regression

```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
```

## data retrieval

```python
data = scipy.io.loadmat("data/logistic_fit.mat")

t: np.ndarray = data["t"].reshape(-1)
y: np.ndarray = data["y"].reshape(-1)
```

## problem setup

```python
a = np.hstack([t.reshape(-1, 1), np.ones((len(t), 1))])
b = np.log(y / (1 - y))
```

## solving and display

```python
alpha, beta = np.linalg.lstsq(a, b)[0]
```

```python
exp_ab = np.exp(alpha * t + beta)
pred_y = exp_ab / (1 + exp_ab)

sorted_t = np.argsort(t)
plt.plot(t[sorted_t], pred_y[sorted_t], label="predicted y", color="orange")
plt.scatter(t, y, label="actual y", marker=".")
plt.legend()
```

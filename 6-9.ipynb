# holy crap, fft?

```python
import numpy as np
import scipy
import time
```

# naive method

```python
n_to_test = [100, 1000, 10000]

for n in n_to_test:
    a = np.random.randn(n)
    b = np.random.randn(n)

    start = time.time()
    # i'd argue that the time it takes to generate the matrix also counts
    circ_a = scipy.linalg.circulant(a)
    res = np.linalg.solve(circ_a, b)
    print(f"naive method on n={n}: {time.time() - start} s")

    abs_err = np.abs((circ_a @ res) - b).sum()
    print(f"total abs error: {abs_err}")
```

# fft method

```python
for n in n_to_test:
    a = np.random.randn(n)
    b = np.random.randn(n)

    start = time.time()
    wa, wb = np.fft.fft(a), np.fft.fft(b)
    diag_inv = 1 / wa
    res = np.fft.ifft(diag_inv * wb)  # numpy does the 1/n for you apparently
    print(f"fft method on n={n}: {time.time() - start} s")

    # i could calculate T(a)x w/ fft as well but i cba
    circ_a = scipy.linalg.circulant(a)
    abs_err = np.abs((circ_a @ res) - b).sum()
    print(f"total abs error: {abs_err}")
```

anyways, yeah, the fft method is orders of magnitude faster for large $n$
